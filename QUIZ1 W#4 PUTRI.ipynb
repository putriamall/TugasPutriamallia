# LOAD DATA

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("apy.csv")
data.head()
```

# DISTRIBUSI SEASON

```python
sns.countplot(data['Season'])
plt.show()
```

# DISTRIBUSI TAHUN HASIL PANEN

```python
data['Crop_Year'].value_counts().plot(kind="pie", autopct="%.2f")
plt.show()
```

```python
pro = data['Production']
plt.hist(pro, bins=5)
plt.show()
```

```python
pro = data['Production']
sns.distplot(pro, bins=5)
plt.show()
```
