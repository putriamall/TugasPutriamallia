```python
string = ""
bar = 1

x = int(input("Masukkan angka :"))

# Looping Baris
while bar <= x:
    kol = bar

    # Looping Kolom
    while kol > 0:
        string = string + " * "
        kol = kol - 1
        
    string = string + "\n"
    bar = bar + 1
print (string)

```

```python
nama = str(input("Masukan Nama :"))

gol = input("Golongan Kerja :")

jam = int(input("Masukan jam Kerja :"))


##Menghitung Golongan Jabatan

if gol == "A" or gol == "a":

    upah =5000*jam;

elif gol == "B" or gol == "b":

    upah =7000*jam;

elif gol == "C" or gol == "c":

    upah =8000*jam;
    
elif gol == "D" or gol == "d":

    upah =10000*jam;

else:

    upah =0;

#menghitung rumus

if jam > 48:

    lembur = jam - 48

    hlembur = lembur * 4000
    
else:
    
    hlembur = 0
    
    print ("tidak mendapat upah lembur")


rumus = upah + hlembur


print ("\n")

print ("===============================================================================")

print ("")

print ("\t\t\t Hitung Gaji Karyawan")

print ("\n")

print ("===============================================================================")

print ("Nama \t\t\t\t: ",nama)

print ("Golongan Kerja \t\t\t: ",gol)

print ("Jumlah Jam Kerja \t\t: ",jam)

print ("Tunjangan Jabatan \t\t: ",upah)

print ("Honor Lembur \t\t\t: ",hlembur)

print ("Total Gaji Dan Tunjangan \t: ",rumus)
```
